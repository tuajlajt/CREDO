## 4. Clinical Safety Measures

- **requires_review=True** hardcoded on all clinical output dataclasses — no autonomous decisions
- **HIPAA audit log** — every data access and inference event logged to `logs/audit.log`
- **PHI de-identification** — all inputs pass through Presidio + DICOM tag stripping before processing
- **Emergency escalation** — critical findings (tension pneumothorax, massive PE, etc.) trigger `log_emergency()` and surfaced as `CriticalAlert` in UI
- **UI acknowledgement** — clinician must explicitly acknowledge AI output before it enters the record
- **Standard disclaimer** — displayed on all AI output cards: "AI-generated — requires clinician review before acting"

## 5. Known Limitations

<!-- Fill in as you discover limitations during evaluation -->

- MedGemma 4B may hallucinate findings not present in images — always verify against raw image
- CXR Foundation embeddings may not generalise well to non-standard positioning
- MedASR performs best on clean audio (16 kHz, minimal background noise)
- TxGemma predictions are for research use only — not validated for clinical pharmacology decisions
- All models have limited validation on paediatric populations
